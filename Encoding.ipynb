**Ordinal Encoding**

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['First','Second','Third','Fourth & Above']])
oe.fit(X_train[['Owner_Type']])
```

```python
X_train['Owner_Type']=oe.transform(X_train[['Owner_Type']])
X_test['Owner_Type']=oe.transform(X_test[['Owner_Type']])
```
